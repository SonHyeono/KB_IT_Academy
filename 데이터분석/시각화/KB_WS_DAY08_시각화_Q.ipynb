### 연령대, 성별 별 유효회원수

```python
# '구분'이 '패스트푸드DT' 인 행만 선택하여 df2에 저장하세요.
df2 = df[df['구분'] == '패스트푸드DT']
df2
```

```python
# '연령대' 별 '유효회원수'를 막대그래프로 그리세요.
# hue 옵션을 '성별'로 설정하고, 제목을 '패스트푸드 드라이브쓰루 회원수'로 설정하세요.

sns.barplot(x="연령대", y='유효회원수', data=df2, hue="성별")
plt.title('패스트푸드 드라이브쓰루 회원수')
plt.show()
```

```python
# df에서 '구분'이 '패스트푸드DT' 인 행만 선택하여 왼쪽 그래프에 그리고, '커피DT' 인 행만 선택하여 오른쪽 그래프에 그리세요.
# 그래프는 위와 동일
plt.figure(figsize = (16,6))
plt.subplot(1,2,1)
sns.barplot(x="연령대", y='유효회원수', data=df[df['구분'] == '패스트푸드DT'], hue="성별")
plt.title('패스트푸드 드라이브쓰루 회원수')
plt.subplot(1,2,2)
sns.barplot(x="연령대", y='유효회원수', data=df[df['구분'] == '커피DT'], hue="성별")
plt.title('패스트푸드 드라이브쓰루 회원수')
plt.show()
```

```python
# for 반복문을 사용하여 각 '구분' 별 그래프를 subplot 하세요.
# 그래프는 위와 동일
plt.figure(figsize = (12,16))

for i,j in enumerate(df['구분'].unique()):
    plt.subplot(3,2,i+1)
    sns.barplot(x="연령대", y='유효회원수', data=df[df['구분'] == j], hue="성별")
    plt.title(j+ " 회원수")
    
plt.show()
```

### 연령대, 드라이브쓰루 별 건당금액

```python
# df에서 '드라이브쓰루', '연령대' 별 '건당금액'의 평균을 df3에 저장하세요.
df3 = df.groupby(["드라이브쓰루", "연령대"])[['건당금액']].mean()
df3
```

```python
# 인덱스를 초기화 하세요. (drop 옵션을 False로 설정)
df3 = df3.reset_index(drop=False)
df3
```

```python
# 코드를 실행하세요.

# 그래프 축 설정
plt.rcParams['axes.spines.top'] = False # 위쪽 축 제거
plt.rcParams['axes.spines.left'] = False # 오른쪽 축 제거
plt.rcParams['axes.spines.right'] = False # 오른쪽 축 제거

# 색깔 설정
colors = sns.color_palette('Paired', 6)
colors1 = [colors[4], colors[0]]
colors2 = [colors[5], colors[1]]
colors
```

```python
# df3에서 '연령대' 별 '건당금액' 을 막대 그래프와 선 그래프로 그리세요.
# 두 그래프 모두 hue 옵션을 '드라이브쓰루'로 설정하고, palette 옵션을 각각 colors1, colors2로 설정하세요.

# 막대 그래프
# 추가 옵션 : alpha = 0.8
plt.rcParams['axes.spines.top'] = False # 위쪽 축 제거

fig, ax = plt.subplots()
ax2 = ax.twinx()

sns.barplot(ax = ax, data = df3, x = '연령대', y = '건당금액', hue = '드라이브쓰루', palette = colors1, alpha = 0.8)
sns.lineplot(ax = ax, data = df3, x = '연령대', y = '건당금액', hue = '드라이브쓰루', palette = colors2, legend = False, style = '드라이브쓰루', markers = ["^", "o"],
            markersize = 12, markeredgewidth = 0.5, dashes = [(1,0), (1,0)])


# 선 그래프
# 추가 옵션 : legend = False, style = '드라이브쓰루', markers = ["^", "o"],
#             markersize = 12, markeredgewidth = 0.5, dashes = [(1,0), (1,0)]


# 제목, 그리드 설정
plt.title('연령대 별 드라이브쓰루 건당 금액', fontsize = 16, y = 1.05)
plt.grid(axis = 'y', color = 'lightgray', linestyle = '--', linewidth = 1.5)
plt.show()
```

### 업종, 세대 별 건당 금액

```python
# df에서 '구분', '세대' 별 '건당금액'의 평균을 df4에 저장하고, 인덱스를 초기화하세요.
df4 = df.groupby(["구분","세대"])["건당금액"].mean().reset_index()
df4
```

```python
# df에서 '구분' 별 '건당금액'의 평균을 df5에 저장하고, 인덱스를 초기화하세요.
df5 = df.groupby("구분")["건당금액"].mean().reset_index()
df5
```

```python
# df4에서 '구분' 별 '건당금액'을 막대 그래프로 그리세요.
# hue 옵션을 '세대'로 설정하고, hue_order 옵션을 ['청소년', '청년', '중년', '노년']로 설정하세요.
# 추가 옵션 : palette = 'hls'

plt.rcParams['axes.spines.top'] = False # 위쪽 축 제거

fig, ax = plt.subplots()
ax2 = ax.twinx()

sns.barplot(ax = ax, data = df4, x = '구분', y = '건당금액', hue = '세대', hue_order =['청소년', '청년', '중년', '노년'] , palette = 'hls')
sns.lineplot(ax = ax, data = df5, x = '구분', y = '건당금액', palette = 'hls', label = '평균', color = 'red', marker = 'o', markersize = 8, markeredgewidth = 0.5)


# df5에서 '구분' 별 '건당금액'을 선 그래프로 그리세요.
# 추가 옵션 : label = '평균', color = 'red', marker = 'o', markersize = 8, markeredgewidth = 0.5

# 제목, 그리드 설정
plt.title('업종 별 건당 금액', fontsize = 16, y = 1.05)
plt.grid(axis = 'y', color = 'lightgray', linestyle = '--', linewidth = 1.5)
plt.show()
```

```python
# df4와 df5에서 '무인스터디카페'를 제외하고 위와 같은 그래프를 그리세요.
df4_temp = df4.drop(df4[(df4['구분'] == '무인스터디카페')].index)
df5_temp = df5.drop(df5[(df5['구분'] == '무인스터디카페')].index)


plt.rcParams['axes.spines.top'] = False # 위쪽 축 제거

fig, ax = plt.subplots()
ax2 = ax.twinx()

sns.barplot(ax = ax, data = df4_temp, x = '구분', y = '건당금액', hue = '세대', hue_order =['청소년', '청년', '중년', '노년'] , palette = 'hls')
sns.lineplot(ax = ax, data = df5_temp, x = '구분', y = '건당금액', palette = 'hls', label = '평균', color = 'red', marker = 'o', markersize = 8, markeredgewidth = 0.5)

plt.title('업종 별 건당 금액', fontsize = 16, y = 1.05)
plt.grid(axis = 'y', color = 'lightgray', linestyle = '--', linewidth = 1.5)
plt.show()
```

### (참고) 20, 30대 이용 현황

```python
# df에서 '세대'가 '청년'인 행만 선택하여 df6에 저장하세요.
df6 = df[df['세대'] == '청년']
df6
```

```python
# df6의 '구분' 별 '유효회원수'의 합, '건당금액'의 평균을 구하세요.
df6 = pd.concat([df.groupby('구분')[['유효회원수']].sum(), df.groupby('구분')[['건당금액']].mean()],axis=1)
df6
```

```python
# 인덱스를 초기화한 후, '유효회원수'를 기준으로 내림차순으로 정렬하세요.
df6 = df6.reset_index()
df6 =df6.sort_values(by='유효회원수', ascending=False)



df6
```

```python
fig, ax = plt.subplots()
ax2 = ax.twinx()

sns.barplot(ax = ax, data = df6,
            x = '구분', y = '유효회원수', palette = 'hls')
sns.lineplot(ax = ax2, data = df6,
            x = '구분', y = '건당금액', label = '건당금액', palette = 'hls',
            color = 'red', marker = 'o', markersize = 8, markeredgewidth = 0.5)

ax2.legend(loc = 'upper center')
plt.title('20, 30대 이용 현황', fontsize = 16, y = 1.05)
plt.grid(axis = 'y', color = 'lightgray', linestyle = '--', linewidth = 1.5)
plt.show()
```
