```python
import pandas as pd
score = pd.DataFrame(
    {
        "이름":["홍","김"],
        "영어":[70, 90],
        "수학":[50, 80],
    }
)
score
```

```python
def calc_avg(x):
    return x.mean()
    
score["평균"] = score[["영어","수학"]].apply(calc_avg, axis=1)
score
```
