```python
from tensorflow.keras import datasets
(x_tr, y_tr),(x_te, y_te) = datasets.imdb.load_data(num_words=10000)
x_tr.shape, x_te.shape
```

```python
word_index = datasets.imdb.get_word_index()
word_index
```

```python
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
reverse_word_index
```

```python
print(x_tr[:2])
```

```python
' '.join([ reverse_word_index.get(i-3, '?') for i in x_tr[0] ])
# 만약 없는 단어를 요청하면 '?'로 대체하겠다.
# imdb가 index_from=3을 기본값으로 가지기에 i-3 해야함.
```

```python
import numpy as np

def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1
    return results
x_tr = vectorize_sequences(x_tr)
x_te = vectorize_sequences(x_te)
x_tr
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers

model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile
```

```python
from tensorflow.keras import models, layers
model = models.Sequential() # 순차처리객체
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(
    x_tr,
    y_tr,
    epochs=20,
    batch_size=512,  # batch_size를 주지않으면 train 데이터를 한꺼번에 읽음
    validation_split=0.2
)
```

```python
model.evaluate(x_te, y_te)
```

```python
import matplotlib.pyplot as plt
plt.plot( history.history['loss'], label = 'loss')
plt.plot( history.history['val_loss'], label = 'val_loss')
plt.legend()
plt.show()
```

```python
# 타이타닉을 NN의 이진분류 적용하기
```

```python
import seaborn as sns
df = sns.load_dataset('titanic')
df = df.loc[:,['age','pclass','sex','survived']]
df
```

```python
df.info()
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
import pandas as pd
df = pd.get_dummies(df, columns=['pclass','sex'])
df
```

```python
X = df.drop('survived', axis=1)
y = df['survived']
```

```python
from sklearn.model_selection import train_test_split

x_tr, x_te, y_tr, y_te = train_test_split(
    X,
    y,
    random_state=0,
    stratify=y
)
```

```python
x_tr.shape, y_tr.shape, x_te.shape, y_te.shape
```

```python
from tensorflow.keras import models, layers
model = models.Sequential() # 순차처리객체
model.add(layers.Dense(64, activation='relu', input_shape=(x_tr.shape[1],)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(
    x_tr,
    y_tr,
    epochs=500,
    verbose=0
)
```

```python
model.evaluate(x_te, y_te)
```

```python
import matplotlib.pyplot as plt
plt.plot( history.history['loss'], label = 'loss')
plt.plot( history.history['accuracy'], label = 'accuracy')
plt.legend()
plt.show()
```
