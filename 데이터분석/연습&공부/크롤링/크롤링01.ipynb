## 환율 파싱

```python
url="https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%ED%99%98%EC%9C%A8"
import requests
r= requests.get(url=url)

# print("r", r.text)

from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, "html.parser")

tr_arr = soup.select("table.rate_table_info > tbody > tr")
print(len(tr_arr))
환율 = dict()
for tr in tr_arr:
    k = tr.select("th")[0].text
    v = tr.select("td")[0].text    
#     문자열은 불변객체( Immutable Object )
    v = v.replace(",", "")
    환율[ k ] = float(v)
print(환율)
```

```python
# 강사님이 힌트 주신 거에서 경기가 없습니다 까지 나오게 했습니다!

url="https://www.kovo.co.kr/game/v-league/11110_schedule_list.asp"
import requests
r = requests.get(url=url)

from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, "html.parser")

tr_arr = soup.select("table.lst_board.lst_schedule > tbody > tr")
print(len(tr_arr))
for tr in tr_arr:
    dt = tr.select_one(".date").text
    game_info = ""
    right = tr.select_one(".tright") 
    if right : # right에 값이 존재하면
        game_info += right.text
    left = tr.select_one(".tleft") 
    if left: # left에 값이 존재하면
        game_info += left.text
    none = tr.select_one(".left")
    if none:
        game_info = none.text
        
    print(dt, game_info)
/html/body/section[2]/article/section/article[2]/div[4]/div/table/tbody/tr[1]
```
