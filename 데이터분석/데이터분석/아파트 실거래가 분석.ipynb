```python
import pandas as pd

apt_price = pd.read_csv(
    "./아파트(매매)__실거래가_20220401113837.csv",
    skiprows= 15,
    encoding="euc-kr",
    thousands=",")
apt_price
```

```python
apt_price["평"] = apt_price["전용면적(㎡)"] / 3.03
apt_price.head()
```

```python
apt_price["계약년"] = apt_price["계약년월"] // 100
apt_price["계약월"] = apt_price["계약년월"] % 100
apt_price.head()
```

```python
apt_price["시군구"].str.split()
```

```python
apt_price["계약년월"] + apt_price["계약일"] # 이렇게 하면 그냥 숫자가 더해진거
```

```python
# str로 타입을 바꿔야지 가능하다.
import numpy as np
apt_price["계약일"] = np.where(apt_price["계약일"] < 10 ,
        "0" + apt_price["계약일"].astype(str), apt_price["계약일"].astype(str) )
apt_price["계약년월일"] = apt_price["계약년월"].astype(str) + apt_price["계약일"].astype(str) 
apt_price["계약년월일"] = pd.to_datetime(apt_price["계약년월일"])
```

```python
apt_price["평형"] = pd.cut(
    apt_price["평"],
    bins=[0,10,20,30,40,50,60,1000], # 구간경계값
    labels=['10평이하', '10평대', '20평대', '30평대', '40평대', '50평대','50평이상']
)
apt_price.head()
```

```python
apt_price["평형"].value_counts()
```

```python
#거래금액을 0~1억, 10억, 20억, 30억, 40억 50억, 50억이상
apt_price["거래금액구간"] = pd.cut(
    apt_price["거래금액(만원)"],
    bins=[0,10000,100000,200000,300000,
          400000,500000,1000000000], # 구간경계값
    labels=[
        '1억이하',
        '1억대',
        '10억대',
        '20억대',
        '30억대',
        '40억대',
        '50억이상']
)
apt_price.head()
```

```python
# 시군구를 분할 ~~시 ~~구 ~~동
apt_price["구"] = apt_price["시군구"].str.split().str[1]
apt_price["동"] = apt_price["시군구"].str.split().str[2]

apt_price["구"]
```

```python
# 20평대 구별 평균 거래금액 출력
cond = apt_price["평형"] =="20평대"
apt_price.loc[cond, :].groupby("구")[["거래금액(만원)"]].mean()\
    .sort_values(by="거래금액(만원)", ascending=False)
```
